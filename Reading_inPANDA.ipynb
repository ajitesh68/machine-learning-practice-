```python
import pandas as pd
from io import StringIO

# JSON formatted as an array
Data = '[{"employee_name":"james","email":"jamesz@gmail.com"}]'

# Use StringIO for a file-like object
df = pd.read_json(StringIO(Data))

print(df)
```
