```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,4,6]
y=[8,7,6,5]

#create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
x=[2,3,4,6]
y=[8,7,6,5]


#create a customized line plot

plt.plot(x,y,color='red',linestyle='-.',marker='*',linewidth=3)
plt.grid()
```

```python
x=[2,3,4,6]
y1=[8,7,6,5]
y2=[3,4,5,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("plot 1")

plt.subplot(1,2,2)
plt.plot(y1,x,color='orange')
plt.title("plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='purple')
plt.title("plot 3")

plt.subplot(2,2,4)
plt.plot(y2,x,color='blue')
plt.title("plot 4")
```

```python
#Bar Plor
categories =['A','B','C','D']
values = [5,6,7,7,]

plt.bar(categories,values,color='purple')
```

```python
#create histogram
data = [3,4,5,7,6,9,8,0,8,7,6,5]

plt.hist(data,bins=5,color='red',edgecolor='black')
```

```python
#create a scatter plot
x=[2,3,4,6]
y=[8,7,6,5]
plt.scatter(x,y,color="orange",marker='x')
```

```python
#pipe chart

labels=['A','B','C','D']
sizes=[30,40,20,10]
colors=['gold','blue','red','white']
explode=(0,2,0,0) ##move out the first slice

#create a pie chart
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```
